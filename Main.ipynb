## Import Libraries / Global Variables

```python
from sklearn.decomposition import PCA
from glob import glob
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

pca = PCA(n_components=2)
np.set_printoptions(precision=2, edgeitems=4)
direc = glob('Data/*')
```

## Read / Transform Data

```python
def img_to_arr(file):
    img = Image.open(file).resize((512, 512)).convert('L') #convert to grey scale to reduce from 3d to 2d
    return np.array(img).flatten()
```

```python
data = np.array([img_to_arr(i) for i in direc])
print("data.shape:\n{}".format(data.shape))
print("------------")
print("data:\n{}".format(data))
```

## Decompose Data

```python
data_d = pca.fit_transform(data)
print("data_d.shape:\n{}".format(data_d.shape))
print("------------")
print("data_d:\n{}".format(data_d))
```

## Visually Inspect New Data

```python
plt.scatter(data_d[:,0], data_d[:,1], s=5)
plt.plot([0], [0], marker='o', markersize=2, color="red") #going to reconstruct point (0, 0)
plt.show()
```

## Reconstruct Data

```python
data_r = pca.inverse_transform(data_d)
print("data_r.shape:\n{}".format(data_r.shape))
print("------------")
print("data_r:\n{}".format(data_r))
```

## Convert Reconstruction to Images

```python
def arr_to_img(arr):
    plt.imshow(np.rot90(arr.reshape(512, 512), 3), cmap='Greys')
    plt.show()
```

```python
arr_to_img(data_r[15])
```

```python
data_00 = pca.inverse_transform(np.array([0,0]))
plt.imshow(np.rot90(data_00.reshape(512, 512), 3), cmap='Accent')
plt.show()
```
